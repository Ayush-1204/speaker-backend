```python
import os
import numpy as np
import soundfile as sf
import librosa
from tqdm import tqdm
from pathlib import Path
from speechbrain.pretrained import EncoderClassifier
from google.colab import drive

# Mount Drive
drive.mount('/content/drive')

# Paths
ROOT = Path('/content/drive/MyDrive/Speaker_Recognition_System')
RAW = ROOT / 'data/raw'
FEATURES = ROOT / 'data/features'
FEATURES.mkdir(parents=True, exist_ok=True)

# Load ECAPA model
model = EncoderClassifier.from_hparams(source="speechbrain/spkrec-ecapa-voxceleb")
print('✅ ECAPA model loaded')
```

```python
def extract_embedding(filepath):
    signal, sr = librosa.load(filepath, sr=16000)
    emb = model.encode_batch(torch.tensor(signal).unsqueeze(0))
    return emb.squeeze(0).detach().cpu().numpy()
```

```python
import torch

metadata = []
X = []
y = []

wav_files = sorted(list(RAW.glob("*.wav")))

print(f'Found {len(wav_files)} audio files')
for file in tqdm(wav_files):
    label = 1 if "familiar" in file.name else 0
    embedding = extract_embedding(file)

    X.append(embedding)
    y.append(label)

    metadata.append({"file": file.name, "label": label})

X = np.array(X)
y = np.array(y)

np.save(FEATURES / "X_embeddings.npy", X)
np.save(FEATURES / "y_labels.npy", y)

import json
with open(FEATURES / "metadata.json", "w") as f:
    json.dump(metadata, f, indent=2)

print("✅ Embeddings saved!")
print("Shape:", X.shape)
```
