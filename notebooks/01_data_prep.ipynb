```python
from google.colab import drive
drive.mount('/content/drive')

from pathlib import Path
ROOT = Path('/content/drive/MyDrive/Speaker_Recognition_System')
DATA_RAW = ROOT / 'data' / 'raw'
DATA_PROCESSED = ROOT / 'data' / 'processed'
FEATURES = ROOT / 'data' / 'features'
METADATA = ROOT / 'metadata'
MODELS = ROOT / 'models'
for d in [DATA_RAW, DATA_PROCESSED, FEATURES, METADATA, MODELS]: d.mkdir(parents=True, exist_ok=True)
print('✅Folders Ready:', ROOT)
```

```python
!pip install --quiet webrtcvad soundfile pydub
import webrtcvad, soundfile as sf, numpy as np
from pydub import AudioSegment
vad = webrtcvad.Vad(3)

def vad_process(audio, sr=16000, frame_ms=30):
    frame_len = int(sr * frame_ms / 1000)
    voiced=[]
    for i in range(0, len(audio)-frame_len, frame_len):
        frame = audio[i:i+frame_len]
        if vad.is_speech((frame*32768).astype(np.int16).tobytes(), sr): voiced.extend(frame)
    return np.array(voiced, dtype=np.float32)
print('✅VAD READY')
```

```python
for wav in DATA_RAW.glob('*.wav'):
    audio, sr = sf.read(wav)
    if audio.ndim>1: audio = audio.mean(axis=1)
    voiced = vad_process(audio)
    if len(voiced)>0:
        sf.write(DATA_PROCESSED/wav.name, voiced, sr)
print('✅Completed — Files:', list(DATA_PROCESSED.glob('*.wav')))
```
